```python
# specify the path of the Excel file
file_path = '../../data/qr_answers.xlsx'
sheet_name = 'Export'  # sheet name

# import the Excel sheet as a pandas DataFrame
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Remove rows with NaN values in the 'Comment' column
df['Antwort'] = df['Antwort'].astype(str)
df.dropna(subset=['Antwort'], inplace=True)
df = df[df['Antwort'].str.len() >= 5]
df['Antwort'] = df['Antwort'].str.replace('\n', ' ')
docs = df["Antwort"].to_list()
```

## Predict

```python
# Load the pre-trained BERTopic model
model = BERTopic.load("../../models/BERTTopic_paraphrase-multilingual-MiniLM-L12-v2_merged.model")

# Use the BERTopic model to predict the topics of the comments
topics, probs = model.transform(docs)
new_topics = model.reduce_outliers(docs, topics) # Reduce outliers
model.update_topics(docs, topics=new_topics) # update Model

# Add the predicted topics to a new column in the dataframe
df['predicted_topic'] = new_topics
```
